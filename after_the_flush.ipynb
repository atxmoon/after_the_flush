## "Afer the Flush" Case Study Observations & Key Challenges for Public Health

In Addis Ababa, Ethiopia, rapid population growth coupled with seasonal flooding and drought have strained existing infrastructures that are already operating beyond maximum capacity. Sanitary system overflow into public streets, cracks in pipeline networks that have been built into mountainous terrain, and intentional damage from unsanctioned access of drinking water lines have facilitated bacterial contamination of water used city-wide for both drinking and agricultural irrigation. 

Contamination is further problematic where both the Kotebe and Kaliti treatment facilities are operating at or beyond maximum capacity, specifically the Kotebe facility which accounts for roughly 70% of Addis Ababa's waste management. Operation of the Kotebe facility requires high human exposure to raw sewage and lacks the technology to properly receive and treat waste that is ultimately returned to natural waterways, causing loss of aquatic biodiversity as well as bacterial outbreaks that impact all groups of the city's population. Children are especially impacted, with diarrhea caused by exposure to pathogens as the leading cause of death for the city’s residents under the age of 5 years old.


<a class="anchor" id="vacuums"></a>
## <center style="background-color:Gainsboro; width:80%;">Vacuum Truck Stem</center>

**Dr. Eyasu's Findings:**
Dr. Eyasu observed each phase behind the logistics of vacuum trucks from collection sites to the Kotebe treatment facility, located at a high elevation and accessible by dirt road. At pit-latrine sites that facilitate roughly 70% of collection in Addis Ababa, Eyasu noted long lines for access to public latrines for a fee, as well as significant disparities in the cost of public versus private vacuum truck services. The Kotebe treatment facility, designed nearly 20 years ago and intended only for sludge drying, was functioning at maximum capacity with diminished effectiveness due to inadequate technology.

**Impact**

• Transfer stations in dense areas significantly compromise public health, more so for children under 5.

• Inadequate facility design is causing biodiversity loss from untreated waste water pollution.

**Strengths**

• Established infrastructure; 3 cubic meter vacuum trucks can navigate otherwise inaccessible tight roads.

**Challenges**

• Capacity; lines to use facilities may encourage alternative solutions, while limited trucks encourage gouging from private companies.

• Technology; lack of treatment facility designed with intent of purpose.

<a class="anchor" id="lines"></a>
## <center style="background-color:Gainsboro; width:80%;">Sewer Lines Stem</center>

**Dr. Eyasu's Findings:**
Eyasu next observed Addis Ababa's existing network of sewer lines which facilitate less than 10% of the city's waste management, primarily due to high cost with little funding. The lines use downward slopes within the city to leverage gravity rather than expensive pumps. Sewage lines often overflowed from manholes, as the city only had 2 working vacuum trucks available to clear any blockages. At the Kaliti treatment facility, Dr. Eyasu found that sanitation processes were well monitored and controlled, and the quality of water re-entering the environment was within the national standard. However, like other waste management resources in Addis Ababa, the Kaliti facility was consistently at or beyond maximum capacity. Any expansion of the sewage line networks would also require treatment facility expansion.

**Impact**

• Inclusion of drying beds at site (to compensate overflow from Kotebe), diminish environmental benefits.

• Overall benefits include improved public health and convenience/ease of use.

**Strengths**

• Established stabilization pond system with reliable retention time, producing efficient results with monitored environmental impact.

**Challenges**

• Cost; mountainous terrain increases cost of both construction and maintenance, while no funding exists from sewer charges.

• Capacity; especially negated during seasonal flooding.

## <center style="background-color:Gainsboro; width:80%;">Condominium Stem</center>

**Dr. Eyasu's Findings:**
Dr. Eyasu found that the decentralized use of membrane bioreactors provided the highest quality waste management of the three strategies observed in Addis Ababa. The membrane bioreactors provide the capacity to meet the demands of a growing population, and are located on site at condominiums that are likewise designed to accommodate long term growth. Additionally, the quality of effluent that is released into the environment after treatment is well below the national standard, and clean enough to be
recycled as greywater for use in construction and irrigation. The critical challenge of this waste management strategy is the cost of technology and adequate funding, currently only supported by the city.

**Impact**

• Membrane bioreactors provide the most environmentally friendly waste management strategy.

**Strengths**

• Efficiency and capacity; other facilities are already beyond maximum capacity, while the decentralized treatment at condominiums can support continued population growth.

• Produces the highest quality effluent.

**Challenges**

• Cost of operation and skilled labor.

• Funding; residents do not currently pay for waste management and the city has limited resources.

## Recommendation Plan<a class="anchor" id="recommendations"></a>

To provide long term, efficient, and environmentally safe management of human waste as the population of Addis Ababa continues to grow over the long term, it is critical that:

**1)** funding is addressed as a shared responsibility, and

**2)** the use of vacuum trucks to transport raw sewage from intermediate pay-per-use collection sites is phased out as public pit-latrines are replaced with membrane bioreactors (MBRs), starting with the most densely populated areas of the city.

## Constraints<a class="anchor" id="constraints"></a>

**Funding Constraint:** Residents of Addis Ababa are not currently charged for the 20% of waste managed either by MBR or central sewage lines, both of which provide greater public health benefits over the transport of waste via vacuum trucks. Although MBRs require high cost for technology and maintenance, these systems provide effluent quality well below the national standard that can be sold back to residents for non-potable uses in irrigation and construction.

**Population Constraint:** Although a short-term population decrease is expected as residents have returned to rural areas during the COVID-19 pandemic, the population is expected to more than double with ~13 million people by the year 2050. As most of the expected fluctuation will be in dense areas with limited sanitation infrastructure, it is critical to leverage any lull in resource consumption in the present to prepare for a healthy future. 

**Political Constraint:** While any business venture or investment made during the immediate years following COVID-19 will come with a certain degree of increased risk and uncertainty, there is little room for argument when it comes to which waste management strategy presents the greatest benefit to public health. Apart from leaving the lightest footprint, membrane bioreactors have proven to have the capacity to manage waste in highly populated areas, whereas any attempt to expand management via centralized sewage line will also require expansion of the Kaliti treatment facility to meet capacity needs.

**Public Health Constraint:** The impact of future pandemics is greatly dictated by the health of the environment around us. As we recover globally and populations return to normalized growth, the implications of environmental health grow as well. Membrane bioreactors not only limit human exposure to harmful pathogens, but also protect the water quality that supports all life surrounding Addis Ababa.
